```python
%matplotlib inline
from qiskit import QuantumCircuit, transpile
from qiskit.circuit.library import QFT
from qiskit.visualization import plot_bloch_multivector
from qiskit_aer import AerSimulator
from utils import get_image_path
import matplotlib.pyplot as plt
```

```python
backend = AerSimulator()
```

```python
qc = QuantumCircuit(3)
qc.append(QFT(3), range(3))
qc.draw('mpl', filename=(get_image_path("QFT_1.png")))
```

```python
plot = plot_bloch_multivector(qc)
plot.savefig(get_image_path("QFT_1_bloch.png"))
plot
```

```python
transpile(qc, backend).draw('mpl', filename=get_image_path("QFT_1_transpiled.png"))
```

```python
qc.append(QFT(3, inverse=True), range(3))
qc.draw('mpl', filename=(get_image_path("QFT_2.png")))
```

```python
plot = plot_bloch_multivector(qc)
plot.savefig(get_image_path("QFT_2_bloch.png"))
plot
```

```python
transpile(qc, backend).draw('mpl', filename=get_image_path("QFT_2_transpiled.png"))
```
